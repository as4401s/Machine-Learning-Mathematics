```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.array([3,1])
v
```

```python
#create a function to plot the vector
#make an arrow from (0,0) to the vector point
def plot_vector2d(vector2d, origin=[0,0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1], **options)
```

```python
#plot
plot_vector2d(v,head_width=0.2,color='r',length_includes_head=True)
plt.axis([-5,5,-5,5])
plt.grid()
plt.show()
```

```python
#multiply with a 2x2 identity vector

I = np.identity(2)
I
```

```python
V = np.dot(v,I)
V
```

```python
#Flip vector v along x axis and plot
E = np.array([[1,0],[0,-1]])

Ev = np.dot(v,E)
```

```python
#plot both vectors
#add arrow in front of the vectors

plt.arrow(0,0,v[0],v[1],head_width=0.2,color='r',length_includes_head=True)
plt.arrow(0,0,Ev[0],Ev[1],head_width=0.2,color='b',length_includes_head=True)
plt.axis([-5,5,-5,5])
plt.grid()
plt.show()
```

```python
#Flip over y axis

F = np.array([[-1,0],[0,1]])

Fv = np.dot(v,F)
```

```python
#plot
plt.arrow(0,0,v[0],v[1],head_width=0.2,color='r',length_includes_head=True)
plt.arrow(0,0,Fv[0],Fv[1],head_width=0.2,color='b',length_includes_head=True)
plt.axis([-5,5,-5,5])
plt.grid()
plt.show()
```

```python
#affine

A = np.array([[-1,4],[2,-2]])

Av = np.dot(v,A)
```

```python
#plot
plt.arrow(0,0,v[0],v[1],head_width=0.2,color='r',length_includes_head=True)
plt.arrow(0,0,Av[0],Av[1],head_width=0.2,color='b',length_includes_head=True)
plt.axis([-5,5,-5,20])
plt.grid()
plt.show()
```
